```python
import pandas as pd

# Load your dataset (replace 'your_data.csv' with your actual data file)
data = pd.read_csv('/kaggle/input/facebook-page-insights-dataset/Apr-19-2022_Jul-18-2022.csv')

# Extract quantitative columns (int64 and float64)
quantitative_data = data.select_dtypes(include=['int64', 'float64'])

# Define a threshold for null values (adjust as needed)
null_threshold = 0.5  # Columns with >50% null values are considered "prominent"

# Calculate the percentage of null values in each column
null_percentage = quantitative_data.isnull().mean()

# Identify columns with null percentages below the threshold
non_prominent_null_columns = null_percentage[null_percentage <= null_threshold].index

# Filter the quantitative data to keep only non-prominent null columns
filtered_quantitative_data = quantitative_data[non_prominent_null_columns]

# Display the quantitative columns with prominent null columns removed
print(filtered_quantitative_data)
```

# Histogram

```python
import matplotlib.pyplot as plt

# Create a histogram for a specific quantitative column
plt.hist(filtered_quantitative_data['Impressions'], bins=20)
plt.title('Histogram of Impressions')
plt.xlabel('Impressions')
plt.ylabel('Frequency')
plt.show()
```

# Box Plot

```python
# Create a box plot for a specific quantitative column
plt.boxplot(filtered_quantitative_data['Likes'])
plt.title('Box Plot of Likes')
plt.ylabel('Likes')
plt.show()
```

# Scatter Plot

```python
# Create a scatter plot for two specific quantitative columns
plt.scatter(filtered_quantitative_data['Likes'], filtered_quantitative_data['Shares'])
plt.title('Scatter Plot of Likes vs. Shares')
plt.xlabel('Likes')
plt.ylabel('Shares')
plt.show()
```

# Pair Plot (using specific columns)

```python
import seaborn as sns

# Create a pair plot for specific quantitative columns
selected_columns = ['Likes', 'Shares', 'Engagements']
sns.pairplot(filtered_quantitative_data[selected_columns])
plt.title('Pair Plot of Selected Columns')
plt.show()
```

# Heatmap (correlation)

```python
# Create a correlation heatmap for all quantitative columns
correlation_matrix = filtered_quantitative_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
